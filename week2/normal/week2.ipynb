# Exercise 1
- ### Calculate the probability of each value in $S$ to be produced by $D$.
Dado que $C$ assume valores de zero até $2^{8}$ (256), significa que temos mais valores, para além do conjunto $S$ (valores de 0 a 250). Por isto, podemos afirmar que irão haver valores, neste conjunto $S$, que se irão repetir. E esses valores são:
$$
251 \mod{251} = 0
$$
$$
252 \mod 251 = 1
$$
$$
253 \mod 251 = 2
$$
$$
254 \mod 251 = 3
$$
$$
255 \mod 251 = 4
$$
Ou seja, o valor 251 vai assumir o valor zero no conjunto $S$, o valor 251 o valor 1, e assim sucessivamente.
Com isto temos que os valores de zero a cinco repetem-se 2 vezes, por isso, vamos ter as seguintes probabilidades:
$$
P(0 \leq X \leq 4) = \frac{2}{251}
$$
$$
P(5 \leq X \leq 250) = \frac{1}{251}
$$
- ### Repeat the above considering now the set $C$ to be the set of all bit strings of length 64.
Considerando agora que $C$ é um conjunto de bit strings de tamanho 8, temos que $C$ assume valores de zero até $2^{64}$.
Dado que o valor obtido é muito grande, a forma que temos para calcular o número de vezes que $C$ vai "colidir" no conjunto $S$ é obtendo o valor inteiro da divisão:
$$
Colisões = \frac{2^{64}}{251}
$$
Sendo que o numerador representar o tamanho de $C$ e o denominador os valores do conjunto $S$ (0..250)

O próximo passo é perceber quais são os valores que sobram. Para isso, fazemos:
$$
2^{64} -1 \mod{251} = 68
$$
O que significa que os valores de zero até 68 vão repetir-se mais uma vez.
Logo, as probabilidades são:
$$
P(0 \leq X \leq 68) = \frac{ Colisões + 1 }{ 251 } = \frac{ \frac{2^{64}}{251} + 1 }{251}
$$
$$
P(69 \leq X \leq 250) = \frac{ Colisões }{251} = \frac{ \frac{2^{64}}{251} }{251}
$$
- ### Are these distributions uniform? If not, can you think of a way to quantify how distant they are from uniform?
As distribuções anteriores não são uniformes.

# Exercise 2
Considerando:
- p = $2^{8}$
- $S$ um conjunto de inteiros no intervalo de 0..255

- ### Calculate the probability of each value in $S$ to be produced by $D$.
Dado que $C$ assume valores de zero até $2^{8}$ (256), significa que todos os valores de $C$ estão dentro do conjunto $S$. Para além disso, o conjunto $S$ será preenchido pelo conjunto $C$, e não haverá repetições, logo a probabilidade é uniforme.
$$
P(0 \leq X \leq 255) = \frac{1}{255}
$$

- ### Repeat the above considering now the set $C$ to be the set of all bit strings of length 64.
Considerando agora que $C$ é um conjunto de bit strings de tamanho 8, temos que $C$ assume valores de zero até $2^{64}$.
$$
Colisões = \frac{2^{64}}{2^{8}} = 2^{56}
$$
Sendo que o numerador representar o tamanho de $C$ e o denominador os valores do conjunto $S$ (0..250)

O próximo passo é perceber quais são os valores que sobram. Para isso, fazemos:
$$
(2^{64}-1) \mod{2^{8}} = 255
$$
O que significa que o ultimo valor de $C$ preencherá o ultimo valor do conjunto $S$. Isto significa que todos os valores de $S$ terão a mesma frequência.
Logo, a probabilidade é:
$$
P(0 \leq X \leq 255) = \frac{ Colisões }{2^{64}} = \frac{ 2^{56} }{2^{64}} = \frac{1}{2^{8}}
$$
- ### Are these distributions uniform? If not, can you think of a way to quantify how distant they are from uniform?
As distribuções anteriores são uniformes.

````python
# Exercise 3
def entropy_prob_i(prob):
    return -prob * math.log2(prob)  

##########################################
## Para os valores $2^{8}$ e $p$ = 251
entropy1 = 0
for _ in range(0, 5):
    entropy1 += entropy_prob_i(2/251)
for _ in range(5, 251):
    entropy1 += entropy_prob_i(1/251)
print(float(entropy1)) # 8 bits

##########################################
## Para os valores $2^{64}$ e $p$ = 251
entropy2 = 0
for _ in range(0, 69):
    entropy2 += entropy_prob_i(2/251)#(2**64 / 251 + 1) / 251)
for _ in range(69, 251):
    entropy2 += entropy_prob_i(1/251)#(2**64 / 251) / 251)
print(float(entropy2)) # 8 bits

##########################################
## Para os valores $2^{8}$ e $p$ = $2^{8}$
entropy3 = 0
for _ in range(0, 2**8):
    entropy3 += entropy_prob_i(1/2**8)
print(float(entropy3)) # 8 bits

##########################################
## Para os valores $2^{64}$ e $p$ = $2^{8}$
entropy4 = 0
for _ in range(0, 2**8):
    entropy4 += entropy_prob_i(1/2**8)
print(float(entropy4)) # 8 bits
````

````python
# Exercise 4
def entropy_prob_i(prob):
    return -prob * math.log2(prob)  

def entropy_d(k, p_val):
    c = 2**k
    colisoes = c // p_val

    res = 0

    # Uniform
    if colisoes == (c/p_val):
        for _ in range(0, p_val):
            prob = 1 / p_val
            res += entropy_prob_i(prob)
    # Not Uniform
    else:
        ultimo_val_repetido = (c % p_val)
        for _ in range(0, ultimo_val_repetido):
            prob = 2 / p_val
            res += entropy_prob_i(prob)
        for _ in range(ultimo_val_repetido, p_val):
            prob = 1 / p_val
            res += entropy_prob_i(prob)
    return res

print(entropy_d(2, 2)) # Moeda
print(entropy_d(8, 251))
print(entropy_d(64, 251))
print(entropy_d(8, 2**8))
print(entropy_d(64, 2**8))
````

# Exercise 5
```
hexdump -n 32 -e '1/4 "%0X" 1 "\n"' /dev/urandom
```
- hexdump: ferramenta que mostra o conteúdo em formato hexadecimal
- -n 32: lê apenas 32 bytes do arquivo de entrada (neste caso o /dev/urandom)
- -e '1/4 "%0X" 1 "\n"': define o formato de saída:
    - 1/4: lê 4 bytes de cada vez.
    - "%0X": imprime o valor lido em hexadecimal maiúsculo, sem prefixo 0x.
    - 1 "\n": imprime 1 unidade por linha, seguida de uma quebra de linha.
- /dev/urandom: fonte de bytes aleatórios.
___
```
dd if=/dev/urandom bs=4 count=8 2>/dev/null | hexdump -e '1/4 "%0X" 1 "\n"'
```
- if=/dev/urandom: entrada de bytes aleatórios.
- bs=4: lê blocos de 4 bytes.
- count=8: lê 8 blocos → 32 bytes no total.
- 2>/dev/null: esconde a saída de estatísticas do dd.
___
```
openssl rand 32 | hexdump -e '1/4 "%0X" 1 "\n"'
```
- openssl rand 32: gera 32 bytes aleatórios.

```
openssl genrsa -aes256 4096
```
- genrsa: gera uma chave privada RSA.
- -aes256: criptografa a chave privada usando AES-256 com uma senha.
- 4096: tamanho da chave em bits

O tamanho da chave, assim como o tempo para criar a chave aumentam proporcionalmente ao tamanho da chave.

Para responder à questão de "Investigate how openSSL converts the passphrase into a cryptography key for encryption/wrapping.", encontramos este site: https://docs.openssl.org/3.3/man7/passphrase-encoding/#pkcs12.

No PKCS#12, OpenSSL codifica senhas como BMPString (UCS-2 Big Endian), adaptando UTF-8 ou ASCII conforme necessário.
Para descriptografar corretamente, é essencial usar a mesma codificação da senha original, e OpenSSL fornece o EVP_KDF-PKCS12KDF para derivação segura de chaves.
