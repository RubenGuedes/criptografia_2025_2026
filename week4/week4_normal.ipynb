### 1.2: Can you ascertain which is the best polynomial for an LFSR?
O melhor polinómio é o $x^{4} + x + 1$, pois, para além do número de bits ser 4, este é **primitivo**, ou seja, temos que o número de periodos máximo que, neste caso, é de 15. Para os restantes polinómios, apenas conseguem gerar 7 estados diferentes.

```python
# 1.3
R.<x> = GF(2)[] # polynomials modulo 2

f1 = x^3 + x + 1
f2 = x^4 + x + 1
f3 = x^4 + x^3 + x^2 + 1

print("Is (x^3 + x + 1) irreducible? :", f1.is_irreducible())
print("Is (x^4 + x + 1) irreducible? :", f2.is_irreducible())
print("Is (x^4 + x^3 + x^2 + 1) irreducible? :", f3.is_irreducible())

print("O terceiro polinómio não está irredutivel, por isso é que vimos no exercício 1.2, que o número de peridos são 7. No entanto, com 4 bits de comprimento, deveria ser possivel gerar 2^4 - 1 = 15 estados possiveis.")
```

Este comportamento é semelhante ao modo AES-CTR, pois cada byte do plaintext é combinado com um byte do keystream, sem necessidade de padding.
